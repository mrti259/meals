```python
import json

# Leemos la data
filepath = "graphql"
with open(filepath, "r") as f:
    print(f"Reading {filepath}")
    data = json.load(f)

# La limpiamos y guardamos
if "chat" in data[0]["data"]:
    print("Deleting chat data")
    del data[0]["data"]["chat"]
    with open(filepath, "w") as f:
        json.dump(data, f)
```

```python
import pandas as pd

# Generamos DF con los datos
recipes = data[0]["data"]["recipes"]
df = pd.json_normalize(recipes)
df
```

```python
df_filtered = (
    df[["title", "labelsId", "images", "nutritionFacts.calories", "preparation.ingredients", "preparation.instructions", "preparation.prepTime"]]
    .rename(lambda col: col.split(".")[-1], axis=1)
)
df_filtered
```

```python
df_sorted = df_filtered.sort_values("calories", ascending=False)
df_sorted
```

```python
with open("recipes.json", "w") as f:
    df_sorted.to_json(f, orient="records")
```
